```python
wine_data = pd.read_csv("/kaggle/input/spanish-wine-quality-dataset/wines_SPA.csv", sep=",")
```

```python
wine_data.head()
```

```python
import seaborn as sns
g = sns.pairplot(data=wine_data, diag_kind="kde", dropna=True)
g.map_lower(sns.kdeplot, levels=4, color=".2")
```
